# Lupton RGB

This notebook illustrates how to create a Lupton RGB image from three monochromatic images.

**NOTE: Uses function from https://github.com/parejkoj/astropy/tree/luptonRGB**

Define the monochromatic images to combine. For now, we use fake random data:

```python
import numpy as np

image_r = np.random.random((100,100))
image_g = np.random.random((100,100))
image_b = np.random.random((100,100))
```

Make the RGB image:

```python
#from astropy.visualization import lupton_rgb  # Future
from mosviz.utils import lupton_rgb

# Additional fileName keyword saves the image as well.
image = lupton_rgb.makeRGB(image_r, image_g, image_b)
```

Display the result:

```python
%matplotlib inline

lupton_rgb.displayRGB(image)
```
